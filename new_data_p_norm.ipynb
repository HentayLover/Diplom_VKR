# Считывает файлы,записывает в новую папку, удаляет лишние столбцы


```python
import os

# Указать путь к папке с файлами
input_folder = r'3ОФ34 - 120mm'
output_folder = 'New_traectory'

# Создать папку для сохранения новых файлов, если её нет
if not os.path.exists(output_folder):
    os.makedirs(output_folder)
    print(f"Папка для сохранения файлов создана: {output_folder}")
else:
    print(f"Файлы будут сохранены в существующую папку: {output_folder}")

# Считывание всех файлов из указанной папки
for filename in os.listdir(input_folder):
    if filename.endswith('.txt'):  # Проверяем расширение файла
        input_path = os.path.join(input_folder, filename)
        output_path = os.path.join(output_folder, filename)

        with open(input_path, 'r', encoding='utf-8') as infile:
            lines = infile.readlines()

        # Обработка строк: берем первые 4 столбца и заменяем запятые на точки
        new_lines = []
        for line in lines:
            columns = line.strip().split()
            if len(columns) >= 4:
                # Обработка только первых 4 столбцов
                processed_line = ' '.join(columns[:4])
                processed_line = processed_line.replace(',', '.')
                new_lines.append(processed_line)

        # Запись результата в новый файл
        with open(output_path, 'w', encoding='utf-8') as outfile:
            outfile.write('\n'.join(new_lines))

print("Обработка завершена.")
```

# Удаляет лишнин строки

```python
import os

# Указать путь к папке New_traectory
output_folder = r'New_traectory'

# Проверка, существует ли папка
if not os.path.exists(output_folder):
    print(f"Папка {output_folder} не существует. Проверьте путь.")
    exit()

# Считывание всех файлов из папки New_traectory
for filename in os.listdir(output_folder):
    if filename.endswith('.txt'):  # Проверяем расширение файла
        file_path = os.path.join(output_folder, filename)

        with open(file_path, 'r', encoding='utf-8') as infile:
            lines = infile.readlines()

        # Удаление первой строки
        lines = lines[1:]

        # Фильтрация строк, оставляем только те, где первый столбец делится на 2
        filtered_lines = []
        for line in lines:
            columns = line.strip().split()
            if len(columns) > 0:
                try:
                    t_value = float(columns[0].replace(',', '.'))  # Преобразуем запятую в точку и парсим как float
                    if t_value % 2 == 0:
                        filtered_lines.append(line)
                except ValueError:
                    print(f"Ошибка преобразования числа в строке: {line.strip()}")

        # Перезапись файла с обновленными данными
        with open(file_path, 'w', encoding='utf-8') as outfile:
            outfile.write(''.join(filtered_lines))

print("Обработка завершена.")
```

# Создание новый файлов

```python
import os
import random

# Указать путь к папке New_traectory
output_folder = r'New_traectory'

# Проверка, существует ли папка
if not os.path.exists(output_folder):
    print(f"Папка {output_folder} не существует. Проверьте путь.")
    exit()

# Функция для получения следующего доступного имени файла
def get_next_filename(base_path, prefix):
    counter = 1
    while True:
        new_filename = f"{prefix}_{counter}.txt"
        new_filepath = os.path.join(base_path, new_filename)
        if not os.path.exists(new_filepath):
            return new_filepath
        counter += 1

# Задать количество повторений алгоритма
iterations = 10000

for _ in range(iterations):
    # Выбор случайного файла из папки
    files = [f for f in os.listdir(output_folder) if f.endswith('.txt')]
    if not files:
        print("Нет файлов для обработки в папке.")
        break

    random_file = random.choice(files)
    file_path = os.path.join(output_folder, random_file)

    # Чтение содержимого файла
    with open(file_path, 'r', encoding='utf-8') as infile:
        lines = infile.readlines()

    # Генерация случайных чисел для всего столбца
    delta_2 = random.uniform(-2000, 2000)
    delta_4 = random.uniform(-2000, 2000)

    # Обработка файла: изменение 2-го и 4-го столбцов
    modified_lines = []
    for line in lines:
        columns = line.strip().split()
        if len(columns) >= 4:
            try:
                # Преобразование и добавление случайных чисел
                columns[1] = str(float(columns[1].replace(",", ".")) + delta_2)
                columns[3] = str(float(columns[3].replace(",", ".")) + delta_4)
                modified_lines.append(' '.join(columns))
            except ValueError:
                print(f"Ошибка преобразования строки: {line.strip()}")
                modified_lines.append(line.strip())

    # Сохранение нового файла с уникальным именем
    new_file_path = get_next_filename(output_folder, "created")
    with open(new_file_path, 'w', encoding='utf-8') as outfile:
        outfile.write('\n'.join(modified_lines))

    print(f"Создан файл: {new_file_path}")

print("Обработка завершена.")
```

## Построение графиков для проверки

```python
import os
import random
import plotly.graph_objects as go

# Указать путь к папке New_traectory
output_folder = r'New_traectory'

# Проверка, существует ли папка
if not os.path.exists(output_folder):
    print(f"Папка {output_folder} не существует. Проверьте путь.")
    exit()

# Поиск файлов, начинающихся на 'created'
created_files = [f for f in os.listdir(output_folder) if f.startswith('created') and f.endswith('.txt')]
if not created_files:
    print("Нет файлов, начинающихся на 'created', для построения графика.")
    exit()

# Выбор случайного файла
random_created_file = random.choice(created_files)
created_file_path = os.path.join(output_folder, random_created_file)

# Чтение данных из файла
x_vals, y_vals, z_vals = [], [], []
with open(created_file_path, 'r', encoding='utf-8') as file:
    for line in file:
        columns = line.strip().split()
        if len(columns) >= 4:
            try:
                x_vals.append(float(columns[1]))
                y_vals.append(float(columns[2]))
                z_vals.append(float(columns[3]))
            except ValueError:
                print(f"Ошибка преобразования строки: {line.strip()}")

# Построение графика с использованием Plotly
fig = go.Figure(data=[go.Scatter3d(
    x=x_vals,
    y=z_vals,
    z=y_vals,
    mode='lines',
    line=dict(width=2),
)])

fig.update_layout(
    title=f"3D График для файла: {random_created_file}",
    scene=dict(
        xaxis_title='X (м)',
        yaxis_title='Z (м)',
        zaxis_title='Y (м)'
    )
)

fig.show()

print(f"3D-график построен для файла: {random_created_file}")
```

# Перезапись в новую систему координат

```python
import os
import random

# Путь к папке с исходными файлами
input_folder = r'New_traectory'

# Путь к папке для сохранения новых файлов
output_folder = r'New_traectory_new_cord'

# Создание выходной папки, если её нет
if not os.path.exists(output_folder):
    os.makedirs(output_folder)
    print(f"Создана папка для новых файлов: {output_folder}")

# Получение списка файлов
files = [f for f in os.listdir(input_folder) if os.path.isfile(os.path.join(input_folder, f))]

for file_name in files:
    input_file_path = os.path.join(input_folder, file_name)

    # Чтение файла
    with open(input_file_path, 'r', encoding='utf-8') as file:
        lines = file.readlines()

    # Проверка, содержит ли файл больше 10 строк
    if len(lines) <= 10:
        continue

    # Выбор случайной строки с учётом минимального количества строк до и после неё
    min_lines_before = 4  # Минимальное количество строк до выбранной
    min_lines_after = 4   # Минимальное количество строк после выбранной
    max_index = len(lines) - min_lines_after - 1  # Индекс, до которого можно выбрать случайную строку
    min_index = min_lines_before  # Индекс, от которого можно выбрать случайную строку
    random_index = random.randint(min_index, max_index)
    random_line = lines[random_index].strip().split()

    # Проверка формата строки
    if len(random_line) < 4:
        print(f"Файл {file_name} содержит некорректную строку: {lines[random_index]}")
        continue

    try:
        # Преобразование значений строки
        t_offset = float(random_line[0])
        x_offset = float(random_line[1])
        z_offset = float(random_line[3])
    except ValueError:
        print(f"Ошибка преобразования строки: {random_line}")
        continue

    # Создание нового содержимого файла
    new_lines = []
    for line in lines:
        columns = line.strip().split()
        if len(columns) >= 4:
            try:
                t = float(columns[0]) - t_offset
                x = float(columns[1]) - x_offset
                y = float(columns[2])  # Третий столбец остается без изменений
                z = float(columns[3]) - z_offset
                new_lines.append(f"{t:.4f} {x:.4f} {y:.4f} {z:.4f}")
            except ValueError:
                print(f"Ошибка обработки строки: {line.strip()}")
                continue

    # Сохранение нового файла
    output_file_path = os.path.join(output_folder, file_name)
    with open(output_file_path, 'w', encoding='utf-8') as output_file:
        output_file.write('\n'.join(new_lines))
    
    print(f"Обработан файл: {file_name}, сохранен в {output_file_path}")

print("Обработка всех файлов завершена.")
```

# Нормировка

```python
import os

# Путь к папке с исходными файлами
input_folder = r'New_traectory_new_cord'

# Путь к папке для сохранения нормализованных файлов
output_folder = r'New_traectory_norm'

# Создание папки для нормализованных файлов, если её нет
if not os.path.exists(output_folder):
    os.makedirs(output_folder)
    print(f"Создана папка для нормализованных файлов: {output_folder}")

# Инициализация значений для максимума
max_t, max_x, max_y, max_z = float('-inf'), float('-inf'), float('-inf'), float('-inf')

# Получение списка файлов
files = [f for f in os.listdir(input_folder) if os.path.isfile(os.path.join(input_folder, f))]

# Первый проход для нахождения максимальных значений
for file_name in files:
    input_file_path = os.path.join(input_folder, file_name)

    # Чтение файла
    with open(input_file_path, 'r', encoding='utf-8') as file:
        lines = file.readlines()

    for line in lines:
        columns = line.strip().split()
        if len(columns) >= 4:
            try:
                # Преобразование значений в float
                t = float(columns[0])
                x = float(columns[1])
                y = float(columns[2])
                z = float(columns[3])

                # Обновление максимальных значений
                max_t = max(max_t, t)
                max_x = max(max_x, x)
                max_y = max(max_y, y)
                max_z = max(max_z, z)

            except ValueError:
                # Если ошибка при преобразовании, пропустить эту строку
                continue

# Проверка, что максимальные значения найдены
if max_t == float('-inf') or max_x == float('-inf') or max_y == float('-inf') or max_z == float('-inf'):
    print("Ошибка: Не удалось найти максимальные значения.")
    exit()

# Вывод максимальных значений
print(f"Максимальное значение t: {max_t}")
print(f"Максимальное значение x: {max_x}")
print(f"Максимальное значение y: {max_y}")
print(f"Максимальное значение z: {max_z}")

# Второй проход для нормализации данных и сохранения файлов
for file_name in files:
    input_file_path = os.path.join(input_folder, file_name)
    normalized_lines = []  # Список для новых строк

    # Чтение файла
    with open(input_file_path, 'r', encoding='utf-8') as file:
        lines = file.readlines()

    for line in lines:
        columns = line.strip().split()
        if len(columns) >= 4:
            try:
                # Преобразование значений в float
                t = float(columns[0]) / max_t
                x = float(columns[1]) / max_x
                y = float(columns[2]) / max_y
                z = float(columns[3]) / max_z

                # Добавление строки в список нормализованных данных
                normalized_lines.append(f"{t:.6f} {x:.6f} {y:.6f} {z:.6f}")
            except ValueError:
                # Если ошибка при преобразовании, пропустить эту строку
                continue

    # Сохранение нормализованных данных в новый файл
    output_file_path = os.path.join(output_folder, file_name)
    with open(output_file_path, 'w', encoding='utf-8') as output_file:
        output_file.write('\n'.join(normalized_lines))
    
    print(f"Файл {file_name} нормализован и сохранён в {output_file_path}")

print("Нормализация всех файлов завершена.")
```
